# GlicoVoz — Extração de features por paciente e técnica

Este notebook usa `get_patient_features` (em [`feature_utils.py`](./feature_utils.py),
na mesma pasta) para, dado o **ID de um paciente** (ex. `P1R_F25`) e a **técnica de
extração de feature** desejada, retornar um `DataFrame` com os metadados do paciente +
as colunas de feature daquela técnica.

Técnicas suportadas (as 3 famílias usadas no projeto):

- `"praat"` — qualidade vocal via Praat/parselmouth (F0, jitter, shimmer, HNR, formantes...)
- `"spectral"` — espectral/timbre via librosa (MFCC, centroide, rolloff, chroma...)
- `"wav2vec"` — embeddings do Wav2Vec2 PT-BR (mean+std pooling)

**Esta pasta (`notebooks/`) é autossuficiente**: `feature_utils.py` só depende de
`pandas`/`pyarrow` e lê os parquets já extraídos em `notebooks/outputs/features/`
(incluída dentro desta própria pasta). Não é preciso instalar `parselmouth`, `librosa`,
`torch`/`transformers` nem ter o pacote `glicovoz` no path — basta copiar/exportar a pasta
`notebooks/` sozinha, sem o resto do repositório. Nenhum áudio é reprocessado aqui; as
features já foram extraídas por `scripts/extract_features.py` e cacheadas em parquet.
